**-Estadistica Inferencial.**
Se encarga de extraer conclusiones sobre una **población** a través de una **muestra**. Se basa en hipótesis, muestreo e investigaciones. A diferencia de la estadística descriptiva, esta busca una proyección a Futuro .

*** - Las hipótesis no tiene porque ser siempre correcta - ***

Intervalos de confianza: Rango de valores que se espera contenga un parámetro poblacional.


Pruebas de hipótesis: Métodos para determinar si hay suficiente evidencia en una muestra para hacer afirmaciones sobre la población.
A traves de un **Modelado estadístico. S**e genera la posible proyección a partir de los resultados obtenidos en las muestras.

Una empresa de streaming mide el tiempo de respuesta de sus servidores cada hora. Recopilan datos de 50 horas:
Calcularon la media: *2.4 segundos.*
Detectaron desviación estándar: *0.5 segundos.*
Conclusión: Identificaron que durante horas pico *(17:00 a 20:00)*, el tiempo promedio aumentó a *3.8 segundos*, lo que llevó a optimizar consultas de base de datos para reducir la latencia.

Se recopilaron los datos durante 2 días y 2 horas acerca de los servidores de una plataforma de streaming. A partir de estos datos, podemos formular  hipótesis  entre las 5:00 p.m. y las 8:00 p.m., las respuestas de los servidores aumentan considerablemente. Esto podría deberse a las posibles razones.

**Primera hipótesis:** El incremento registrado dentro de la medición pudo haberse provocado por un resultado de una mala medición o de alguna interferencia generada ajena al servidor

**Segunda hipótesis:** La latencia generada pudo ser provocada por uno o varios eventos transmitidos de manera simultánea, lo que trae como consecuencia una sobrecarga mediante la alta demanda del servicio de streaming.

**Tercera hipótesis:** La infraestructura del servidor no esta lo suficientemente suministrada para soporta un alta demanda de los usuarios, lo que se puede deducir como bases de datos lentas y/o filas largas de procesamiento por parte del servidor.

**Generalizando:**  El incremento de la latencia registrado en la medición no necesariamente es el mismo a la misma hora todos los días, ya que la medición resulta ser solo un amuestra de todos los eventos posibles que pueden generar este incremento en la velocidad de respuesta.

Con esta mejora se espera que el tiempo se reduzca, incluso que sea mejor que el tiempo promedio que es de 2.4 segundos en las horas menos activas.
